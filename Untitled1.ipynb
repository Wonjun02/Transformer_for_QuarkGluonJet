```python
#[셀 1] – 모델 및 유틸리티 정의 (CustomSwinTransformer, Forward_features 추가, ProjectionHead, MoCoSwin)
# 셀 1: 모델 및 유틸리티 정의

import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
from PIL import Image
from torchvision import transforms
from torch.utils.data import Dataset

###############################
# Helper functions for window partition
###############################

def window_partition(x, window_size): # 입력 이미지를 윈도우로 분할하는 전처리 함수
    # x: (B, H, W, C) 배치크기, 높이, 너비 채널 수
    B, H, W, C = x.shape
    x = x.view(B, H // window_size, window_size, W // window_size, window_size, C)
    windows = x.permute(0, 1, 3, 2, 4, 5).contiguous().view(-1, window_size, window_size, C) #(N, ,windeow_size, windeow_size, C)
    # N: 전체 윈도우 수
    return windows

def window_reverse(windows, window_size, H, W): # 위에 있는 함수의 역함수(?)
    B = int(windows.shape[0] / (H * W / window_size / window_size))
    x = windows.view(B, H // window_size, W // window_size, window_size, window_size, -1)
    x = x.permute(0, 1, 3, 2, 4, 5).contiguous().view(B, H, W, -1)
    return x

###############################
# Patch Embedding
###############################

class PatchEmbed(nn.Module): # 이미지를 작은 패치들로 분할한 후, 각 패치를 벡터로 임베딩.
    def __init__(self, img_size=66, patch_size=2, in_chans=3, embed_dim=64):
        super().__init__()
        self.proj = nn.Conv2d(in_chans, embed_dim, kernel_size=patch_size, stride=patch_size)
        # 커널 사이즈와 스트라이드가 patch_size로 설정되어 있음 + 출력 채널 수는 emebed_dim
        self.num_patches = (img_size // patch_size) * (img_size // patch_size)
    def forward(self, x):
        # x: (B, 3, 66, 66)
        x = self.proj(x)  # (B, embed_dim, 33, 33)
        x = x.flatten(2).transpose(1,2)  # (B, 33*33, embed_dim)
        return x

###############################
# Swin V2 Window Attention
###############################

class WindowAttentionV2(nn.Module):
    def __init__(self, dim, window_size=3, num_heads=2):
        super().__init__()
        self.dim = dim
        self.window_size = window_size  # 3x3
        self.num_heads = num_heads
        head_dim = dim // num_heads
        self.logit_scale = nn.Parameter(torch.log(torch.tensor(10.0))) # 어텐션 점수를 스케일링 하기 위한 파라미터, 초깃값은 log(10)
        self.qkv = nn.Linear(dim, dim * 3, bias=True) # qkv 선형 변환
        self.proj = nn.Linear(dim, dim) # 어텐션 연산 후의 결과를 다시 원래 차원으로 변환함.
        self.relative_position_bias_table = nn.Parameter( # 상대 위치 편향
            torch.zeros((2 * window_size - 1) ** 2, num_heads) # 윈도우의 가능한 상대 위치 수(2W-1)^2
        )
        coords_h = torch.arange(window_size)
        coords_w = torch.arange(window_size)
        coords = torch.stack(torch.meshgrid([coords_h, coords_w], indexing='ij'))
        coords_flatten = torch.flatten(coords, 1)
        # 좌표간 차이를 계산해서 relative_coords를 얻음.
        relative_coords = coords_flatten[:, :, None] - coords_flatten[:, None, :]
        relative_coords = relative_coords.permute(1,2,0).contiguous()
        relative_coords[:,:,0] += window_size - 1
        relative_coords[:,:,1] += window_size - 1
        relative_coords[:,:,0] *= 2 * window_size - 1
        relative_position_index = relative_coords.sum(-1)
        self.register_buffer("relative_position_index", relative_position_index) # 최종적으로 학습 과정 중 고정된 상대 위치 인덱스를 버퍼로 등록함.
        nn.init.trunc_normal_(self.relative_position_bias_table, std=0.02) # 상대 위치 편향 테이블의 값을 truncated normal 분포로 초기화함.
    def forward(self, x, mask=None):
        B_, N, C = x.shape
        # 선형 레이어를 통해 입력을 Query,Key, Value로 변환 후, 각 헤드별로 토큰들을 분리. 최종적으로 q, k, v 각각의
        # shape는 (B_, num_heads, N, head_dim)
        qkv = self.qkv(x).reshape(B_, N, 3, self.num_heads, C // self.num_heads).permute(2,0,3,1,4)
        q, k, v = qkv[0], qkv[1], qkv[2]
        # Querry와 Key 벡터들을 정규화하여 내적 연산 시 안정성을 높임.
        q = F.normalize(q, dim=-1)
        k = F.normalize(k, dim=-1)
        # 어텐션 점수 계산
        attn = (q @ k.transpose(-2, -1)) # Query와 Key의 내적을 통해 어텐션 점수를 계산함.
        scale = self.logit_scale.exp().clamp(max=100) # 학습 가능한 logit_scale 파라미터를 지수화한 값을 곱해 어텐션 점수를 스케일링 함.
        attn = attn * scale
        # 상대 위치 편향 추가
        # 상대적 위치 인덱스를 사용해 편향 값을 가져온 후, 이를 어텐션 점수에 더햐여 공간적 상대 위치 정보를 반영함.
        relative_position_bias = self.relative_position_bias_table[self.relative_position_index.view(-1)].view(
            self.window_size * self.window_size, self.window_size * self.window_size, -1
        ).permute(2,0,1).unsqueeze(0)
        attn = attn + relative_position_bias
        # 만약 mask가 제공되면 윈도우 개수에 맞춰 어텐션 텐서를 재배열하고, 주어진 마스크를 더해 특정 위치의 어텐션 점수를 제한함.
        if mask is not None:
            nW = mask.shape[0]
            attn = attn.view(B_ // nW, nW, self.num_heads, N, N) + mask.unsqueeze(1).unsqueeze(0)
            attn = attn.view(-1, self.num_heads, N, N)
            attn = F.softmax(attn, dim=-1)
        else:
            attn = F.softmax(attn, dim=-1)
        # 어텐션 결과
        x = (attn @ v).transpose(1,2).reshape(B_, N, C) # 어텐션 행렬과 Value를 곱해 각 토큰에 대한 가중합을 계산함.
        x = self.proj(x) # 선형 레이어를 통해 최종 출력 feature를 변환함. x shape (B_, N, C)
        return x

###############################
# Swin Transformer Block (Swin V2)
###############################

class SwinTransformerBlock(nn.Module):
    def __init__(self, dim, input_resolution, num_heads, window_size=3, shift_size=1, mlp_ratio=4.0):
        # input_resolution: 입력 feature map의 공간적 해상도(H, W)
        super().__init__()
        self.dim = dim
        self.input_resolution = input_resolution
        self.window_size = window_size
        self.shift_size = shift_size
        # 정규화 -> 어텐션 -> 정규화 -> MLP
        self.norm1 = nn.LayerNorm(dim)
        self.attn = WindowAttentionV2(dim, window_size=window_size, num_heads=num_heads)
        self.norm2 = nn.LayerNorm(dim)
        mlp_hidden_dim = int(dim * mlp_ratio)
        self.mlp = nn.Sequential(
            nn.Linear(dim, mlp_hidden_dim),
            nn.GELU(),
            nn.Linear(mlp_hidden_dim, dim)
        )
        H, W = self.input_resolution
        # 원래 해상도(H, W)에서 윈도우 크기로 나누어 떨아지도록 크기를 조정
        H_new = (H // window_size) * window_size
        W_new = (W // window_size) * window_size
        # (1, H_new, W_new, 1) 크기의 텐서로, 모든 값을 0으로 초기화한 후 각 윈도우에 고유한 번호(cnt)를 할당함.
        # 각 슬라이스 영역에 순차적으로 번호를 부여함.
        img_mask = torch.zeros((1, H_new, W_new, 1))
        h_slices = (slice(0, -window_size),
                    slice(-window_size, -shift_size),
                    slice(-shift_size, None))
        w_slices = (slice(0, -window_size),
                    slice(-window_size, -shift_size),
                    slice(-shift_size, None))
        cnt = 0
        for h in h_slices:
            for w in w_slices:
                img_mask[:, h, w, :] = cnt
                cnt += 1
        # 윈도우 분할 및 어텐션 마스크 계산.
        mask_windows = window_partition(img_mask, window_size) # img_mask를 윈도우 단위로 분할함.
        mask_windows = mask_windows.view(-1, window_size * window_size)  # 분할된 윈도우 flatten
        attn_mask = mask_windows.unsqueeze(1) - mask_windows.unsqueeze(2) # 윈도우 내 각 토큰 차이를 계산해서 같은 윈도우는0, 다른 경우에는 값이 다르게 나오도록 설정
        attn_mask = attn_mask.masked_fill(attn_mask != 0, float(-100.0)).masked_fill(attn_mask == 0, float(0.0)) # 0이 아닌 값은 -100으로 채워서 sofrmax 계산 시 마스크 효과를 줌
        self.register_buffer("attn_mask", attn_mask) # 어텐션 마스크를 고정 버퍼로 등록함.
    def forward(self, x):
        B, L, C = x.shape
        H, W = self.input_resolution
        assert L == H * W, "입력 feature의 크기가 잘못되었습니다."
        shortcut = x # 잔차 연결을 위해 원본 입력을 shortcut에 저장함.
        x = self.norm1(x)
        x = x.view(B, H, W, C)
        # 윈도우 시프트 적용
        if self.shift_size > 0:
            shifted_x = torch.roll(x, shifts=(-self.shift_size, -self.shift_size), dims=(1,2))
        else:
            shifted_x = x
        # 윈도우 분할 및 self-attention 수행
        H_new = (H // self.window_size) * self.window_size
        W_new = (W // self.window_size) * self.window_size
        shifted_x = shifted_x[:, :H_new, :W_new, :].contiguous() # 조정된 해상도를 기준으로 shifted_x를 자름
        x_windows = window_partition(shifted_x, self.window_size) # window_partition 함수를 통해 각 윈도우로 분할함.
        x_windows = x_windows.view(-1, self.window_size * self.window_size, C)
        attn_windows = self.attn(x_windows, mask=self.attn_mask) # 각 윈도우를 WindowAttentionV2 모듈에 전달하며, 이전에 생성한 attn_mask를사용해 어텐션 계산을 수행함.
        attn_windows = attn_windows.view(-1, self.window_size, self.window_size, C)
        shifted_x = window_reverse(attn_windows, self.window_size, H_new, W_new) # 분할된 윈도우들을 원래의 공간적 배열로 복원함
        # 만약 시프트를 적용한 경우, 원래 위치로 복원.
        if self.shift_size > 0:
            x = torch.roll(shifted_x, shifts=(self.shift_size, self.shift_size), dims=(1,2))
        else:
            x = shifted_x
        x = x.view(B, H_new * W_new, C)
        x = shortcut[:, :H_new * W_new, :] + x # 잔차 연결
        x = x + self.mlp(self.norm2(x)) # 두 번째 정규화 및 MLP
        return x

###############################
# Patch Merging
###############################

class PatchMerging(nn.Module):
    def __init__(self, input_resolution, dim):
        super().__init__()
        self.input_resolution = input_resolution
        self.reduction = nn.Linear(4 * dim, 2 * dim, bias=False) # 4배가 된 채널을 2배 차원으로 축소하는 선형 변환 레이어를 정의함.
        self.norm = nn.LayerNorm(4 * dim)
    def forward(self, x):
        H, W = self.input_resolution
        B, L, C = x.shape
        assert L == H * W, "입력 feature의 크기가 잘못되었습니다."
        x = x.view(B, H, W, C)
        # 해상도 조정
        H_new = H - (H % 2)
        W_new = W - (W % 2)
        x = x[:, :H_new, :W_new, :]
        # 2x2 패치 추출
        x0 = x[:, 0::2, 0::2, :]
        x1 = x[:, 1::2, 0::2, :]
        x2 = x[:, 0::2, 1::2, :]
        x3 = x[:, 1::2, 1::2, :]
        x = torch.cat([x0, x1, x2, x3], -1)
        x = x.view(B, -1, 4 * C)
        x = self.norm(x)
        x = self.reduction(x)
        return x

###############################
# Stage Module
###############################

# 여러 개의 Swin Transformer Block을 하나의 stage로 묶어, 입력 feature를 순차적으로 처리함.
class SwinTransformerStage(nn.Module):
    def __init__(self, dim, input_resolution, depth, num_heads, window_size=3, shift_size=1):
        super().__init__()
        # Modulelist를 사용하여 여러 개의 Swin Transformer Block들을 순서대로 저장함.
        self.blocks = nn.ModuleList([
            SwinTransformerBlock(dim, input_resolution, num_heads, window_size, shift_size if (i % 2 == 1) else 0)
            for i in range(depth)
        ])
    def forward(self, x): # 입력 x가 각 Swin Transformer Block을 순차적으로 통과하면서 점차 feature가 변환됨.
        for blk in self.blocks:
            x = blk(x)
        return x

###############################
# CustomSwinTransformer (Swin V2)
# forward_features 메서드를 추가하여 classification head 전의 feature를 추출
###############################

class CustomSwinTransformer(nn.Module):
    def __init__(self, img_size=66, patch_size=2, in_chans=3,
                 stage_blocks=[2, 2, 4], stage_dims=[64, 128, 256],
                 stage_heads=[2, 4, 8], window_size=3, shift_size=1, num_classes=2):
        super().__init__()
        self.patch_embed = PatchEmbed(img_size, patch_size, in_chans, stage_dims[0])
        resolution = (img_size // patch_size, img_size // patch_size)  # (33,33)
        # stage1
        self.stage1 = SwinTransformerStage(stage_dims[0], resolution, stage_blocks[0], stage_heads[0], window_size, shift_size)
        # stage2
        self.proj_stage2 = nn.Linear(stage_dims[0], stage_dims[1]) # stage1의 출력을 stage2에 맞춰서 64 -> 128 차원으로 선형 변환.
        self.stage2 = SwinTransformerStage(stage_dims[1], resolution, stage_blocks[1], stage_heads[1], window_size, shift_size)
        self.patch_merging = PatchMerging(resolution, stage_dims[1])
        new_resolution = (15, 15)  # (16,16)에서 truncate하여 (15,15)
        # stage3
        self.stage3 = SwinTransformerStage(stage_dims[2], new_resolution, stage_blocks[2], stage_heads[2], window_size, shift_size)
        self.norm = nn.LayerNorm(stage_dims[2])
        self.head = nn.Linear(stage_dims[2], num_classes)

    def forward_features(self, x):
        x = self.patch_embed(x)         # (B, 1089, 64)
        x = self.stage1(x)              # (B, 1089, 64)
        x = self.proj_stage2(x)          # (B, 1089, 128)
        x = self.stage2(x)              # (B, 1089, 128)
        x = self.patch_merging(x)         # (B, 256, 256)
        x = x[:, :15*15, :]             # (B, 225, 256)
        x = self.stage3(x)              # (B, 225, 256)
        x = self.norm(x)
        x = x.mean(dim=1)              # Global average pooling → (B, 256)
        return x
    def forward(self, x):
        x = self.forward_features(x) # 이미지에서 전역적인 feature 벡터를 추출함.
        x = self.head(x) # 추출된 feature 벡터에 대해 분류 헤드를 적용하여 최종 클래스 logits를 생성함.
        return x

###############################
# Projection Head for MoCo (추가 MLP)
###############################

# moco에서 encoder의 출력 feature를 낮은 차원의 임베딩으로 변환하기 위해 사용됨.
class ProjectionHead(nn.Module):
    def __init__(self, in_dim, hidden_dim=256, out_dim=128):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(in_dim, hidden_dim), # 입력 차원(in_dim)을 hidden_dim으로 선형 변환함.
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, out_dim) # hidden_dim을 최종 임베딩 차원(out_dim)으로 변환함.
        )
    def forward(self, x):
        return self.mlp(x)

###############################
# MoCoSwin: MoCo 스타일 wrapper for Swin encoder(V3는 아님)
###############################
# 두 개의 encoder를 사용하여, 동일한 이미지의 두 가지 augmentation을 각각 'query'와 'key'로 처리함.
# 이때, query와 key의 표현(embedding)이 서로 가깝고, 다른 이미지의 표현은 멀어지도록 contrastive loss를 적용함.
class MoCoSwin(nn.Module):
    def __init__(self, base_encoder, feature_dim=128, m=0.99, T=0.07, queue_size=1024):
        """
        base_encoder: CustomSwinTransformer (classification head는 사용하지 않고 forward_features를 이용)
        feature_dim: Projection head 출력 차원
        m: 모멘텀 계수
        T: temperature
        """
        super().__init__()
        self.m = m # m=0.99: key encoder 업데이트 시, 이전 key encoder 값을 99% 유지하고 query encoder의 업데이트 1%를 반영함
        self.T = T # T=0.07: 어텐션 점수(내적 값)을 temperature로 스케일링하여 loss계산 시 분포를 조절함.
        self.queue_size = queue_size

        # query, key encoder (dummy classification head 사용)
        self.encoder_q = base_encoder(num_classes=2)
        self.encoder_k = base_encoder(num_classes=2)

        # Projection heads (Swin encoder의 global average pooling 출력 차원은 256로 가정)
        self.proj_q = ProjectionHead(in_dim=256, hidden_dim=256, out_dim=feature_dim)
        self.proj_k = ProjectionHead(in_dim=256, hidden_dim=256, out_dim=feature_dim)

        # 초기화: key encoder와 proj_k를 query와 동일하게 초기화하고 grad 계산 없이 업데이트
        for param_q, param_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            param_k.data.copy_(param_q.data)
            param_k.requires_grad = False
        for param_q, param_k in zip(self.proj_q.parameters(), self.proj_k.parameters()):
            param_k.data.copy_(param_q.data)
            param_k.requires_grad = False

        # negative queue: (feature_dim, queue_size) 이전 배치에서 얻은 key feature들을 저장하는 큐를 사용하여
        # 많은 negative 샘플(다른 이미지)와 비교할 수 있게 함.
        # queue: negative 샘플을 저장하는 버퍼로, 임의의 값으로 초기화한 후 정규화함.
        self.register_buffer("queue", torch.randn(feature_dim, queue_size))
        self.queue = F.normalize(self.queue, dim=0)
        # queue_ptr: 큐의 현재 위치를 가리키는 포인터로 사용됨.
        self.register_buffer("queue_ptr", torch.zeros(1, dtype=torch.long))

    # 모멘텀 업데이트
    # key encoder와 proj_k의 파라미터를, qeury encoder와 proj_q의 파라미터를 일정 비율로 업데이트함.
    # 급격환 변화 없이 안정적으로 업데이트 하기 위함.
    @torch.no_grad()
    def _momentum_update_key_encoder(self):
        for param_q, param_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            param_k.data = param_k.data * self.m + param_q.data * (1. - self.m)
        for param_q, param_k in zip(self.proj_q.parameters(), self.proj_k.parameters()):
            param_k.data = param_k.data * self.m + param_q.data * (1. - self.m)

    # Negative Queue 업데이트
    # 현재 배치에서 계산된 key feature들을 negative queue에 추가함.
    # 큐는 FIFO 방식으로 관리되며, 새로운 key feature가 들어오면 오래된 feature를 대체함.
    # 많은 negative 샘플을 활용해 contrastive loss를 계산할 수 있게 해, 모델이 더 다양한 비교를 할 수 있도록 함.
    @torch.no_grad()
    def _dequeue_and_enqueue(self, keys):
        batch_size = keys.shape[0]
        ptr = int(self.queue_ptr)
        if ptr + batch_size <= self.queue_size:
            self.queue[:, ptr:ptr+batch_size] = keys.T
            ptr = (ptr + batch_size) % self.queue_size
        else:
            first_part = self.queue_size - ptr
            self.queue[:, ptr:] = keys[:first_part].T
            second_part = batch_size - first_part
            self.queue[:, :second_part] = keys[first_part:].T
            ptr = second_part
        self.queue_ptr[0] = ptr

    def forward(self, im_q, im_k):
        # Query branch
        # 입력 이미지 im_q를 query encoder를 통해 feature를 추출하고, projectio head를 통해 임베딩을 만듬
        # 결과를 정규화하여 유사도 계산에 사용
        q_feat = self.encoder_q.forward_features(im_q)  # (B, 256)
        q = self.proj_q(q_feat)  # (B, feature_dim)
        q = F.normalize(q, dim=1)

        # Key branch (모멘텀 업데이트 후)
        # 모멘텀 업데이트 후, im_k(같은 이미지의 다른 augmetation)을 key encoder를 통해 처리함.
        with torch.no_grad():
            self._momentum_update_key_encoder()
            k_feat = self.encoder_k.forward_features(im_k)  # (B, 256)
            k = self.proj_k(k_feat)
            k = F.normalize(k, dim=1)

        # logits 계산
        # positive logits
        l_pos = torch.einsum('nc,nc->n', [q, k]).unsqueeze(-1) # query와 key 사이의 내적을 계산하여, 같은 이미지의 두 변형 간의 유사도를 구함.
        # negative logits
        l_neg = torch.mm(q, self.queue.clone().detach()) # query와 queue에 저장된 negative key feature들과의 내적을 계산함.

        # postice와 negative 로짓을 결합하여 하나의 텐서를 만들고, T로 스케일 조정.
        # 정답 레이블은 항상 0으로 설정되어 첫 번째 위치(positive 쌍)가 정답임을 나타냄.
        logits = torch.cat([l_pos, l_neg], dim=1)  # (B, 1+queue_size)
        logits /= self.T
        labels = torch.zeros(logits.shape[0], dtype=torch.long).to(logits.device)

        self._dequeue_and_enqueue(k)

        return logits, labels

print("모델 및 유틸리티 정의가 완료되었습니다.")
```
